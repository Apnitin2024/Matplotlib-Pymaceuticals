## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_group=mouse_df.groupby(['Drug Regimen'])
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
TV_mean=drug_group['Tumor Volume (mm3)'].mean()
TV_median=drug_group['Tumor Volume (mm3)'].median()
TV_var=drug_group['Tumor Volume (mm3)'].var()
TV_sd=drug_group['Tumor Volume (mm3)'].std()
TV_sem=drug_group['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary DataFrame.
drug_stats=pd.merge(TV_mean, TV_median, on='Drug Regimen')
drug_stats.rename(columns={drug_stats.columns[0]:'Mean Tumor Volume',drug_stats.columns[1]:'Median Tumor Volume'}, inplace=True)
drug_stats=pd.merge(drug_stats, TV_var, on='Drug Regimen')
drug_stats.rename(columns={drug_stats.columns[2]:'Tumor Volume Varience'}, inplace=True)
drug_stats=pd.merge(drug_stats,TV_sd, on='Drug Regimen')
drug_stats.rename(columns={drug_stats.columns[3]:'Tumor Volume Std. Dev.'}, inplace=True)
drug_stats=pd.merge(drug_stats,TV_sem, on='Drug Regimen')
drug_stats.rename(columns={drug_stats.columns[4]:'Tumor Volume Std. Err.'}, inplace=True)
drug_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
drug_advStats=drug_group.agg({'Tumor Volume (mm3)':['mean', 'median', 'var', 'std', 'sem']})
drug_advStats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
final4=['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
mouse_df_final4=mouse_df[mouse_df['Drug Regimen'].isin(final4)]

# Start by getting the last (greatest) timepoint for each mouse
mouse_ID_group=mouse_df_final4.groupby('Mouse ID')
last_timepoint=mouse_ID_group['Timepoint'].max()
last_timepoint_df=pd.DataFrame(last_timepoint)

Capomulin_df = mouse_df.loc[mouse_df["Drug Regimen"] == "Capomulin",:]
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
Final_TumorVolume_df=pd.merge(mouse_df, last_timepoint_df, on=['Mouse ID', 'Timepoint'], how='right')
Final_TumorVolume_df=Final_TumorVolume_df[['Mouse ID', 'Drug Regimen', 'Tumor Volume (mm3)']]
Final_TumorVolume_df.reset_index(drop=True, inplace=True)
Final_TumorVolume_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
final4=['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
Cap_Tumor_Vol=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers.
# Locate the rows which contain mice on each drug and get the tumor volumes
for drug in final4:
    volume=Final_TumorVolume_df.loc[Final_TumorVolume_df['Drug Regimen']==drug]
# add subset
    Cap_Tumor_Vol.append(volume['Tumor Volume (mm3)'])

# Determine outliers using upper and lower bounds
for drug in final4:
    tumor_volume=Cap_Tumor_Vol[final4.index(drug)]
    quartiles = tumor_volume.quantile([0.25, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    IQR = upperq-lowerq
    
    lower_bound = lowerq - (1.5*IQR)
    upper_bound = upperq + (1.5*IQR)
    
    outliers = tumor_volume[(tumor_volume < lower_bound) | (tumor_volume > upper_bound)]
    
    print(f"{drug} Potential Outliers: {outliers}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
fig,ax=plt.subplots()
ax.boxplot(Cap_Tumor_Vol, 0,'rs')
ax.set_ylabel('Final Tumor Volume (mm3)')
ax.set_xticklabels(final4)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
Capomulin=mouse_df.loc[mouse_df['Drug Regimen']=='Capomulin']
Mouse_group=Capomulin.groupby('Mouse ID')
average=Mouse_group['Tumor Volume (mm3)'].mean()
Capomulin=pd.merge(Capomulin, average, on='Mouse ID')
x_axis=Capomulin['Weight (g)']
y_axis=Capomulin['Tumor Volume (mm3)_y']
plt.scatter(x_axis, y_axis)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
